```python
import os
import subprocess

env = os.environ.copy()
env["TRANSFORMERS_NO_TF"] = "1"
env["USE_TF"] = "0"

persistent_dir = "/mnt/batch/tasks/shared/LS_root/mounts/clusters/test2/code/Users/filip.trhlik/bias-bench/"
train_file = os.path.join(persistent_dir, "data/text/wikipedia-10.txt")

seeds = [0]

experiments = [
    (
        "run_mlm.py",
        "bert-base-uncased",
        8,
        16,
        os.path.join(persistent_dir, "experiments/run_mlm.py")
    ),
]

for script_name, model_name, bs, ga, script_path in experiments:
    for seed in seeds:
        exp_id = f"dropout_c-{model_name}_s-{seed}"
        out_dir = os.path.join(persistent_dir, "checkpoints", exp_id)

        if os.path.isdir(out_dir):
            continue

        os.makedirs(out_dir, exist_ok=True)
        log_path = os.path.join(persistent_dir, "log.txt")

        cmd = [
            "python", script_path,
            "--model_name_or_path", model_name,
            "--do_train",
            "--train_file", train_file,
            "--max_steps", "2000",
            "--per_device_train_batch_size", str(bs),
            "--gradient_accumulation_steps", str(ga),
            "--max_seq_length", "512",
            "--save_steps", "500",
            "--preprocessing_num_workers", "4",
            "--dropout_debias",
            "--seed", str(seed),
            "--output_dir", out_dir,
            "--cache_dir", os.path.join(persistent_dir, "hf_cache")
        ]

        with open(log_path, "w", encoding="utf-8") as logf:
            logf.write(f"=== LAUNCH {exp_id} ===\n\n")
            process = subprocess.Popen(
                cmd,
                env=env,
                stdout=logf,
                stderr=logf,
                text=True,
                bufsize=1
            )
            try:
                retcode = process.wait()
            except KeyboardInterrupt:
                retcode = process.poll() or -1
                logf.write("\n\n=== INTERRUPTED BY USER ===\n")
                continue
```
